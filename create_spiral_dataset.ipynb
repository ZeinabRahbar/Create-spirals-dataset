```python
import numpy as np
import matplotlib.pyplot as plt

def generate_spiral_dataset(n_samples, n_classes, noise=0.1):
    angle = 2 * np.pi / n_classes
    data = []
    labels = []
    
    for class_idx in range(n_classes):
        radius = np.linspace(0.1, 1, num=n_samples//n_classes)
        theta = np.linspace(class_idx * angle, (class_idx + 1) * angle, num=n_samples//n_classes) + np.random.randn(n_samples//n_classes) * noise
        x = radius * np.cos(theta)
        y = radius * np.sin(theta)
        
        class_data = np.column_stack((x, y))
        data.append(class_data)
        labels.extend([class_idx] * (n_samples//n_classes))
    
    data = np.concatenate(data, axis=0)
    labels = np.array(labels)
    
    return data, labels

# Generate spiral dataset with 500 samples and 3 classes
n_samples = 500
n_classes = 3
data, labels = generate_spiral_dataset(n_samples, n_classes)

# Plot the generated dataset
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title('Spiral Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_spiral_dataset(n_samples, scale_factor, noise=0.1):
    t = np.linspace(0, 2 * np.pi, num=n_samples)
    spiral_1 = np.column_stack((scale_factor * t * np.cos(t), scale_factor * t * np.sin(t)))
    spiral_2 = np.column_stack((-scale_factor * t * np.cos(t), -scale_factor * t * np.sin(t)))
    
    data_modality_1 = spiral_1 + np.random.randn(n_samples, 2) * noise
    data_modality_2 = spiral_2 + np.random.randn(n_samples, 2) * noise
    
    labels = np.concatenate((np.zeros(n_samples), np.ones(n_samples)))
    data = np.concatenate((data_modality_1, data_modality_2))
    
    return data, labels

# Generate spiral dataset with 2000 samples per modality
n_samples = 2000
scale_factor = 0.5
data, labels = generate_spiral_dataset(n_samples, scale_factor)

# Plot the generated dataset
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title('Spiral Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
